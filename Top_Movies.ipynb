### Տվյալների մշակում

Տևողության արտահայտումը րոպեներով

```python
# պետք է աշխատացնել մեկ անգամ
def time_to_mins(s):
  l = s.split(' ')
  if len(l) == 1 and int(l[0][0]) < 4:
    return int(l[0][0]) * 60
  elif len(l) == 1:
    return int(l[0][:-1])
  else:
    return int(l[0][0]) * 60  + int(l[1][:-1])
for i in range(len(final_df_imdb['length'].values)):
  final_df_imdb['length'].values[i] = time_to_mins(final_df_imdb['length'].values[i])


for i in range(len(final_df_imdb['rate'].values)):
  final_df_imdb['rate'].values[i] = float(final_df_imdb['rate'].values[i])
type(final_df_imdb['rate'].values[0])
final_df_imdb['length']
```

 box_gross  և box_budget սյունակները բաժանում ենք 1000000

```python
final_df_imdb["box_gross"] = final_df_imdb["box_gross"] / 1000000
final_df_imdb["box_budget"] = final_df_imdb["box_budget"] / 1000000
final_df_imdb[["box_gross","box_budget"]]
```

Ստեղծում ենք նոր profit(եկամուտ) սյունյակ

```python
final_df_imdb["profit"] = final_df_imdb["box_gross"] - final_df_imdb["box_budget"]
```

```python
final_df_imdb['country']
single_country_all = []
for i in final_df_imdb['country']:
  single_country_all.extend(i.split("\n"))
single_country = pd.Series(single_country_all).value_counts().head(7)
group_country = final_df_imdb['country'].value_counts().head(7)

fig, ax = plt.subplots(figsize=(13, 10))

size = 0.35
ax.pie(group_country, radius=1,  autopct='%1.1f%%', startangle=90,
       colors=["#f9d679", "#6d3f29", "#56625c", "#729468", "#b3c9d8", "#b9d4b2", "#fceab8"],
       wedgeprops=dict(width=size, edgecolor='w', linewidth=6), textprops={'fontsize':16}, pctdistance=0.84)

ax.pie(single_country, radius=1-size,  autopct='%1.1f%%', startangle=90,
       colors=["#82a790", "#99a782", "#b0d978", "#d4ebb6", "#82a7a3", "#a78299", "#815b73"],
        wedgeprops=dict(width=size, edgecolor='lightgray', linewidth=6), textprops={'fontsize':15}, pctdistance=0.8)
leb = []
leb.extend(group_country.index)
leb.extend(single_country.index)
ax.set(aspect="equal", title='Countries')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
# ax.legend(loc='upper right', fontsize='10', labels=leb,)
plt.show()
```
